**one hot encoder**
- one hot encoder name says  at a time one will on and ther will be off
- for example case_statues has two labels 
   - certifird
   -denied
- when you apply one hot encoding on case_statues, it creates two more extra cplumns
   - case_status_certified
   - case_status_denied
   
   
|case_status|case_status_certified|case_status_denied
|-|-|-|
|certified|1|0
|denied|0|1

**Advantage**
- When you develop ML model it is very impportent the columns should be independent each other
- so here case_status creating two extra columns
- Which are independent to each other, which means the row values at a time only one column has 1
- columns are eachother
- which means 90 degree phase shift
- which means perpendicular to each other
- which means orthogonal to each other

**Disadvantage**
- The disadvantage is id a column has 100 unique labels , 100 new columns will be created
- That data will become spare,which means huge
- columns are more more, dimentions are more
- the procesing time more
- The memory consumption more
- curse of dimensionality
