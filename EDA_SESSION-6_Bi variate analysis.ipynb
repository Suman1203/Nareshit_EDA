```python
path=r"C:\Users\veeranna\DATA SCIENCE\Data_sets\Visadataset.xlsx"
visa_df=pd.read_excel(path)
unique_lable=visa_df["continent"].unique()
print(unique_lable)
l1,l2=[],[]
for i in unique_lable:
    c1=visa_df["continent"]==i
    c2=visa_df["case_status"]=="Certified"
    c3=visa_df["case_status"]=="Denied"
    cer_con=c1&c2
    den_con=c1&c3
    l1.append(len(visa_df[cer_con]))
    l2.append(len(visa_df[den_con]))
print(l1)
d1=pd.DataFrame(zip(unique_lable,l1,l2),columns=["continent","certified","Denied"])    
d1
```
