step-5
# adding new column

```python
name=['Ramesh','Suresh','Satish']
age=[30,35,40]
data=zip(name,age)
ind=['A','B','C']
col=['name','age']
df=pd.DataFrame(data,columns=col,index=ind)
df
```

- If you want to add new column

- df['new column']

- you need to have a list which is having some elements

- And that elements need to be equal to number of rows

- city_name=['Hyd','Blr','Chennai']

- def['city']=city_name

```python
city_name=['Hyd','Blr','Chennai']
df['City']=city_name
df
```

**step-6**
# Update the existing column

- if youu want to creat new columnm or update new column
- both are done same way

```python
df['Name']=['Swamy','Asif','Sathwik']
df['name']=['Swamy','Asif','Sathwik']
df
```

**step-7**
# drop the column

- In order to drop the column

- we need to use drop method

- it takes 3 parameters

    - drop column or row
    
    - if column then mention column name
    
    - asix
        - axis=1 referance as column
        
        - axis=0 referance as row
        
    - you want to create a new dataframe or
    
    - you want to overwrite the existing dataframe
    
        - inplace=True

```python
df.drop('Name',axis=1,inplace=True)
df
```

**step-8**
# Drop row

```python
df.drop('C',axis=0,inplace=True)
```

**step-9**
# Save the data frame

```python
df.to_csv("output.csv")
#while saving index is considered as extra column
#df.to_excel("output.xlsx")
```

```python
#read the output csv
pd.read_csv("output.csv")
```
